```python
from scipy.integrate import odeint
import numpy as np
import pylab

def f(y, t):
  return y

y0 = 1
a = 0 # intervalo de integração
b = 2

t = np.arange(a, b, 0.01) # discretizando t

# Solução numérica
y = odeint(f, y0, t)

# Plotando o resultado
pylab.plot(t, y)
pylab.xlabel('t'); pylab.ylabel('y(t)')
```

**EXERCÍCIO 7**

```python
from scipy.integrate import odeint
import numpy as np
import pylab
import matplotlib.pyplot as plt

def f(v, t):
  w = (2000 - v)/(200 - t)
  return w

# COndição inicial e intervalo de integração
y0 = 0
a = 0
b = 5

t = np.arange(a,b,0.01)
y = odeint(f, y0, t)

# Plot da função por EDO
plt.plot(t, y)
plt.xlabel('t(s)'); plt.ylabel('v(m/s)')
plt.title('Solução por EDO')
plt.show()

# Plot da solução original
sol = 10*t-(t**2)/40
plt.plot(t, sol)
plt.xlabel('t(s)'); plt.ylabel('v(m/s)')
plt.title('Solução original')
plt.show()

# Comparando as tangentes
print('Tangente de y: {}'.format(y[5]/5))
print('Tangente da função original: {}'.format(sol[5]/5))
```

**EXERCÍCIO 8**

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# Definindo as constantes
k = 2*pow(10, -6)
m = 100000

def f(y, t):
  return k*(m-y)*y

# Condição inicial e intervalo de integração
y0 = 1000
a = 0
b = 30

t = np.arange(a,b,0.1)

y = odeint(f, y0, t)

# Plot da solução
plt.plot(t, y)
plt.xlabel('Dias'); plt.ylabel('Infectados')
plt.title('Pessoas infectadas em função do tempo')
plt.show()
```
